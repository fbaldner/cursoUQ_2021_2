```python
# Pacotes utilizados
import numpy as np
import statistics as stat
import scipy.stats
import matplotlib.pyplot as plt

# Constantes
a = 2
b = 5
Ns = 1000

# Cálculo dos parâmetros estatísticos:
X      = np.random.beta(a,b,Ns)
mu     = stat.mean(X)
sigma2 = stat.variance(X)
sigma  = stat.stdev(X)
gamma1 = scipy.stats.skew(X)
beta2  = scipy.stats.kurtosis(X)

print(f'A média das amostras é mu = {mu}')
print(f'A variância das amostras é sigma^2 = {sigma2}')
print(f'O desvio padrão das amostras é sigma = {sigma}')
print(f'A assimetria das amostras é gamma1 = {gamma1}')
print(f'A curtose das amostras é beta2 = {beta2}')

# Geração dos números aleatórios.
# Lista de métodos: https://numpy.org/doc/1.16/reference/generated/numpy.random.RandomState.html
# Documentação do pacote statistics: https://docs.python.org/3/library/statistics.html
# Documentação do pacote scipy.stats: https://docs.scipy.org/doc/scipy/reference/stats.html
```
