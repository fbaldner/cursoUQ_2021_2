```python
import numpy as np
import statistics as sts
import matplotlib.pyplot as plt

a  = 5.0
b  = 1.0
Ns = 1000
Pc = 95

X     = np.random.gamma(a,b,Ns)
mu    = sts.mean(X)
sigma = sts.stdev(X)
mu50  = sts.median(X)
mu_upp = mu + sigma
mu_low = mu - sigma

r_plus  = 0.5*(100 + Pc)
r_minus = 0.5*(100 - Pc)

X_upp = np.percentile(X , r_plus)
X_low = np.percentile(X , r_minus)

plt.plot(X , 'xc')
plt.hlines(mu     , 1 , Ns , label='Mean'             , colors='k' , linestyles='solid')
plt.hlines(mu50   , 1 , Ns , label='Median'           , colors='r' , linestyles='dashed')
plt.hlines(mu_upp , 1 , Ns , label='Mean $\pm$ std'   , colors='b' , linestyles='dashdot')
plt.hlines(mu_low , 1 , Ns ,                            colors='b' , linestyles='dashdot')
plt.hlines(X_upp  , 1 , Ns , label='95% interval'     , colors='m' , linestyles='dotted')
plt.hlines(X_low  , 1 , Ns ,                            colors='m' , linestyles='dotted')
plt.legend()
```
